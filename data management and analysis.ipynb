```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sns

#created a dataset!
student_data={"Name":["Karan malhotra","Rishi Senghal","Nishi meher","Samreen khan","Ishan seth"],
              "Phone_number":["9654872135","8541236544","8541526544","9651236544","9121236544"],
              "Gmail":["karan@gmail.com","rishi@gmail.com","nishi@gmail.com","samreen@gmail.com","ishan@gmail.com"],
             "Department":["computer science","information technology","computer science","civil","mechanical"],
             "Year_of_admission":["2019","2018","2019","2022","2022"],
             "Graduation_year":["2023","2022","2023","2026","2026"]}
sd=pd.DataFrame(student_data)

#function to perform deletion,updation,and adding new data!
def fun():
    cond=True
    while cond:
        print("Enter values")
        value=int(input("""1:To see the table records of a student.
               2:To add the new data in table student record.
               3:To delete existing data present in student record.
               4:To update existing data in student record. 
               5:To analyse the existing data in student record.
               6:To exit"""))
        if value==1:
            display(sd)
        elif value==2:
            name=input("enter name of the student")
            pnum=input("enter phone number")
            gmail=input("enter gmail of a student")
            dept=input("enter department name")
            yoa=input("enter year of admission")
            s1={"Name":name,"Phone_number":pnum,"Gmail":gmail,"Department":dept, "Year_of_admission":yoa,"Graduation_year":int(yoa)+4}
            sd=sd.append(s1,ignore_index=True)
            print("Data added succesfully!!")
        elif value==3:
            data=int(input("""enter name or year of the students to delete their data
                     1: to delete the data of the student by their name
                     2: to delete data of the students who are passout of our college """))
            if data==1:
                name=input("enter name of the student")
                sd["Name"]==name
                sd.drop(sd[sd["Name"]==name].index,inplace=True)
                print("Data deleted succesfully!!")
            else:
                year=input("enter year of admission")
                sd.drop(sd[sd["Year_of_admission"]==year].index,inplace =True)
                print("Data deleted succesfully!!")
        elif value==4:
            data=int(input("""enter the data you want to update
                         1: To update name of a student
                         2: to update phonenumber of a student
                         3: To update gmail of a student
                         4: To update department name of a student
                         5: To update name,phonenumber,gmail of a student"""))
            if data==1:
                mailid=input("enter the gmail id of a student to update their name")
                newname=input("enter new name")
                sd.loc[sd.Gmail==mailid,"Name"]=newname
                print("Name updated succesfully")
                display(sd)
            elif data==2:
                mailid=input("enter the gmail id of a student to update their phone number")
                newnum=input("enter new number")
                sd.loc[sd.Gmail==mailid,"Phone_number"]=newnum
                print("number updated succesfully")
                display(sd)
            elif data==3:
                mailid=input("enter the gmail id of a student to update their phone number")
                newmail=input("enter new gmail ")
                sd.loc[sd.Gmail==mailid,"Gmail"]=newmail
                print("gmail updated succesfully")
                display(sd)
            elif data==4:
                mailid=input("enter the gmail id of a student to update their phone number")
                newdept=input("enter new department name")
                sd.loc[sd.Gmail==mailid,"Department"]=newdept
                print("department name updated succesfully")
                display(sd)
            elif data==5:
                mailid=input("enter the gmail id of a student to update their phone number")
                newname=input("enter new name")
                newnum=input("enter new number")
                newmail=input("enter new gmail")
                sd.loc[sd.Gmail==mailid,"Gmail"]=newmail
                sd.loc[sd.Gmail==mailid,"Name"]=newname
                sd.loc[sd.Gmail==mailid,"Phone_number"]=newnum
                print("data updated succesfully")
                display(sd)

        elif value==5:
            plt.figure(figsize=(15, 6))
            sns.countplot(data = sd, x = "Year_of_admission")
            plt.show()

        else :
            cond=False
```
